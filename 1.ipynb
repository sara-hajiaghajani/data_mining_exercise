### MinMaxScaler

```python
T=MinMaxScaler(feature_range=(0, 1))
df_MinMax = pd.DataFrame(T.fit_transform(df2.drop(['International Visitors_A', 'International Visitors_B', 'International Visitors_C'], axis=1 )))
df_MinMax = df_MinMax.rename(columns={0: 'Population growth', 1: 'Total population', 2: 'Area (sq. km)', 3: 'Coronavirus Cases'})
df_MinMax
```

```python
T=Normalizer()
df_normalizer=pd.DataFrame(T.fit_transform(df2.drop(['International Visitors_A', 'International Visitors_B', 'International Visitors_C'], axis=1 )))
df_normalizer= df_normalizer.rename(columns={0: 'Population growth', 1: 'Total population', 2: 'Area (sq. km)', 3: 'Coronavirus Cases'})
df_normalizer
```

### StandardScaler

```python
T=StandardScaler()
df_standard=pd.DataFrame(T.fit_transform(df2.drop(['International Visitors_A', 'International Visitors_B', 'International Visitors_C'], axis=1 )))
df_standard = df_standard.rename(columns={0: 'Population growth', 1: 'Total population', 2: 'Area (sq. km)', 3: 'Coronavirus Cases'})
df_standard
```

```python
df_dummy = df2.iloc[:,4:7]
df_dummy
```

```python
df_standard = df_standard.join(df_dummy)
df_standard = df_standard.rename(columns={0: 'Population growth', 1: 'Total population', 2: 'Area (sq. km)', 3: 'Coronavirus Cases'})
df_standard
```

```python
y_standard=df_standard.iloc[:,3]
y_standard
```

```python
X_standard=df_standard.drop(['Coronavirus Cases'], axis=1)
X_standard
```

# outlier

### z-score

```python
from scipy import stats 
z=np.abs(stats.zscore(df_standard))
z
```

```python
print(np.where(z>3))
```

```python
z[0][2]
```

### IQR Score

```python
Q1=df_standard.quantile(0.25)
Q1
```

```python
Q3=df_standard.quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
(df_standard<Q1-1.5*IQR)|(df_standard>Q3+1.5*IQR)
```

### Removing

```python
df_zScore=df_standard[(z<3).all(axis=1)]
df_zScore.shape
```

```python
df_IQR=df_standard[((df_standard>=Q1-1.5*IQR)&(df_standard<=Q3+1.5*IQR)).all(axis=1)]
df_IQR.shape
```

# Multiple Regression (power 2)

```python
DF=df_zScore[['Total population']]
DF['Total population**2'] = df_zScore['Total population']**2
DF
```

```python
y=df_zScore.iloc[:,3]
y
```

### Creating and Trainung the model

```python
X_train=DF
y_train=y
```

```python
lm2=LinearRegression()
lm2.fit(X_train, y_train)
```

```python
lm2.intercept_
```

```python
lm2.coef_
```

```python
coef_df_new=pd.DataFrame(lm2.coef_, DF.columns, columns=['Coefficient'])
coef_df_new
```

### Prediction from our model

```python
Predict_train=lm2.predict(X_train)
plt.scatter(y_train,Predict_train)
```

### Regression Evaluation metrics

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_train,Predict_train))
print('MSE:', metrics.mean_squared_error(y_train,Predict_train))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train,Predict_train)))
```

```python
DF1 = DF.join(y)
DF1
```
