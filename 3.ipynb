```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('arrhythmia.data', header=None)
df
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer=KNNImputer()
df=pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
df
```

```python
sns.countplot(x=279,data=df)
```

# Standardize the variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scaled_features=scalar.fit_transform(df.drop(279, axis=1))
scaled_features
```

```python
df_normalize=pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_normalize
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(scaled_features,df[279],test_size=0.3)
```

# Using KNN


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics 
from sklearn.metrics import classification_report,confusion_matrix
```

### for n_neighbors=1

```python
knn1=KNeighborsClassifier(n_neighbors=1,metric='euclidean') 
knn1.fit(X_train,y_train)
y_pred_test1=knn1.predict(X_test)
y_pred_train1=knn1.predict(X_train)
```

### confusion matrix and classification report for test data

```python
print(confusion_matrix(y_test,y_pred_test1))
```

```python
 print(classification_report(y_test,y_pred_test1))
    
```

### confusion matrix and classification report for train data

```python
print(confusion_matrix(y_train,y_pred_train1))
```

```python
 print(classification_report(y_train,y_pred_train1))
```

### for n_neighbors=30

```python
knn30=KNeighborsClassifier(n_neighbors=30,metric='euclidean') 
knn30.fit(X_train, y_train)
y_pred_test30=knn30.predict(X_test)
y_pred_train30=knn30.predict(X_train)
```

### confusion matrix and classification report for test data

```python
print(confusion_matrix(y_test,y_pred_test30))
```

```python
 print(classification_report(y_test,y_pred_test30))
```

### confusion matrix and classification report for train data

```python
print(confusion_matrix(y_train,y_pred_train30))
```

```python
print(classification_report(y_train,y_pred_train30))
```

# Gridsearch

```python
 from sklearn.model_selection import GridSearchCV
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_predict=knn.predict(X_test)
```

```python
param_grid={'n_neighbors':list(range(1,31)),'metric':['euclidean','manhattan','cosine']}
```

```python
grid=GridSearchCV(knn,param_grid,verbose=3,cv=5,scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_prediction=grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_prediction))
```

```python
print(classification_report(y_test,grid_prediction))
```
